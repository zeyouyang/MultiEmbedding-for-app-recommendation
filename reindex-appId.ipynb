### 先前檔案中appId因清理後未重新排序，故在此檔案重新排序後存回原檔

```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm, trange
import re
import torch

train_appId = torch.load('./data/google.train.appId.pt') 
val_appId = torch.load('./data/google.val.appId.pt')  
test_appId = torch.load('./data/google.test.appId.pt')  

train_appId = train_appId.view(-1)
val_appId = val_appId.view(-1)
test_appId = test_appId.view(-1)

concat_train_val = torch.cat((train_appId, val_appId), 0)
concat_all = torch.cat((concat_train_val, test_appId), 0) 
concat_all.shape
```

```python
print(train_appId.shape)
print(val_appId.shape)
print(test_appId.shape)
```

```python
concat_all_np = concat_all.numpy()
```

```python
iids = sorted(np.unique(concat_all_np))
item2id = dict(zip(iids, range(0, len(iids))))
test = np.array([item2id[i] for i in concat_all_np])
test
```

```python
item2id
```

```python
new_train_appId = test[0:904120]
new_val_appId = test[904120:1004577]
new_test_appId = test[1004577:]
```

```python
len(new_test_appId)
```

```python
new_train_appId = torch.tensor(new_train_appId, dtype=torch.float32)
new_val_appId = torch.tensor(new_val_appId, dtype=torch.float32)
new_test_appId = torch.tensor(new_test_appId, dtype=torch.float32)

new_train_appId = torch.unsqueeze(new_train_appId,dim=1) 
torch.save(new_train_appId, './data/google.train.appId.pt')

new_val_appId = torch.unsqueeze(new_val_appId,dim=1) 
torch.save(new_val_appId, './data/google.val.appId.pt')  

new_test_appId = torch.unsqueeze(new_test_appId,dim=1) 
torch.save(new_test_appId, './data/google.test.appId.pt')  
```
